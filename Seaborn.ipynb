## Seaborn


## Distribution plots


Univariate analysis - Analysing a feature (distplot
Bi-variate analysis - Analysing realtionship between two features (jointplot)
Multi-variate analysis - Analyzing relationship between three or more features(pairplot)

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

'tip' is the dependent feature, all other features are independent.

```python
df.corr()
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr(),annot=True)
```

## Bivariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

## Pair plot

Multvariate analysis

```python
sns.pairplot(df, hue='sex')
```

## Distplot

helps to check the distributions of the columns feature


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False, bins=10)
```

## Categorical Plots

- boxplot
- violinplot
- countplot
- bar plot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='day',data=df)
```

## Boxplots

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.barplot(x='total_bill',y='sex', data=df)
```

```python
sns.barplot(y='total_bill',x='day', data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=df)
```

## Violin Plot

Helps us to see the distribution both in terms of Kernel density estimation and the boxplot

```python
sns.violinplot(x='total_bill',y='smoker',data=df, palette='rainbow')
```
